```python
from car_blur import find_car_on_image_and_blur, find_car_on_image
from utils import plot_results
from glob import glob
import numpy as np
from tqdm.notebook import tqdm

imgs = glob('../data/car-segmentation/images/*.png')
masks = []
empty = 0
loop = tqdm(imgs)

for IMG_PATH in loop:
    results = find_car_on_image(IMG_PATH)
    masks.append(results['mask'])
    if results['mask'] is None:
        empty += 1
    loop.set_postfix({'no car': empty})
```

# Metrics

```python
from PIL import Image

true_masks = []

for mask in tqdm(glob('../data/car-segmentation/masks/*.png')):
    true_masks.append((np.array(Image.open(mask)) != 0).astype(int))
```

```python
df['true_mask'] = true_masks
df
```

```python
# accuracy
df.apply(lambda x: (x['mask'] == x['true_mask']).mean(), axis=1).mean()
```

```python
# IoU
df.apply(lambda x: (x['mask'] & x['true_mask']).sum() / (x['mask'] | x['true_mask']).sum(), axis=1).mean()
```
